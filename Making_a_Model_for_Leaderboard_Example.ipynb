```python
import pandas as pd
train_df = pd.read_csv('Housing_Data_Train.csv')
```

```python
train_df
```

```python
#Do a linear fit using only the LotArea
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
X = train_df[['LotArea']]
y = train_df['SalePrice']

model = LinearRegression()
model.fit(X,y) #this fits the model
```

```python
#plot the predictions of the model on the training data
train_df = pd.read_csv('Housing_Data_Train.csv')
import matplotlib.pyplot as plt
plt.xlabel("LotArea")
plt.ylabel("SalePrice")
X = train_df[['LotArea']]
y = train_df['SalePrice']
predictions = model.predict(X)
plt.plot(X,y,'b.',label="True SalePrice")
plt.plot(X,predictions,'r.',label="Predicted SalePrice")
plt.legend()
```

```python
#Predict on the test set
test_df = pd.read_csv('Housing_Data_Test.csv')
X = test_df[['LotArea']]
predictions = model.predict(X)

#convert the predictions to a LIST of POSITIVE INTEGERS (do not submit decimals!)
list_predictions = predictions.tolist()
rounded_list_predictions = [int(x) for x in list_predictions]
print(rounded_list_predictions)
```
