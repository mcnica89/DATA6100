```r
#Data setup
myvars <- c("Sepal.Length","Sepal.Width","Petal.Length","Petal.Width")
dat <- iris[myvars]
Species <- iris$Species
summary(dat)
```

```r
library(ggplot2)
#Some scatter plots!
ggplot(dat)+geom_point(aes(x=Sepal.Length,y=Sepal.Width, color=Species)) 
ggplot(dat)+geom_point(aes(x=Petal.Length,y=Petal.Width, color=Species))
```

```r
#Do some PCA
PCA <- prcomp(dat, center=TRUE, scale.=TRUE)
summary(PCA)
```

```r
#Add the PCA as columns to our data
dat <- cbind(dat,PCA$x[,1:4])
summary(dat)
```

```r

#Plots with the PCA features
ggplot(dat)+geom_point(aes(x=PC1,y=PC2, color=Species)) 
ggplot(dat)+geom_point(aes(x=PC3,y=PC4, color=Species)) 
```
